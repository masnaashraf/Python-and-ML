```python
#for loading inbuilt dataset
from sklearn import datasets
iris=datasets.load_iris()
```

```python
import pandas as pd
```

```python
iris
```

```python
#inbulit dataset contain data and target,now  getting data
df=pd.DataFrame(iris.data,columns=iris.feature_names)
#including the column name
```

```python
df
```

```python
#taking target variable
y=iris.target
y
```

```python
print(df.shape)
print(y.shape)
```

```python
#applying decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dec_classifier=DecisionTreeClassifier()
dec_classifier.fit(df,y)
```

```python
#printing decision tree formed using pydotplus
import six
import sys
sys.modules['sklearn.externals.six'] = six
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

dot_data=StringIO()
export_graphviz(dec_classifier, out_file=dot_data,filled=True,feature_names = iris.feature_names, class_names = iris.target_names)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue()) 
Image(graph.create_png())
```
