```python
from sklearn.impute import SimpleImputer
simp_imp=SimpleImputer(strategy="most_frequent")
```

```python

df_categ=pd.DataFrame(simp_imp.fit_transform(categ),columns=categ.columns)
df_categ.head()
```

```python
df_categ.isna().sum()
```

```python
car_data1=[df_new,df_categ]
```

```python
print(car_data1)
```

```python
car_data_com=pd.concat(car_data1,axis=1,join="inner")
car_data_com.head()
```

```python
car_data_com.info()
```

```python
car_data_com["Loan_Status"].unique()
```

```python
x = car_data_com.iloc[:,:-1].values
y = car_data_com.iloc[:,-1].values
```

```python
print(x)
```

```python
x.shape
```

```python
y.shape
```

```python
x=pd.DataFrame(x)
```

```python
#encoding categorical data

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[5,6,7,8,9,10])],remainder='passthrough')
x = np.array(ct.fit_transform(x))
print(x)
```

```python
y=pd.DataFrame(y)
```

```python
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder
dep=LabelEncoder()
y=dep.fit_transform(y)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
x_train = st.fit_transform(x_train)
x_test = st.fit_transform(x_test)
```

```python
#fit the model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 2)
classifier.fit(x_train,y_train)
```

```python
y_pred_train = classifier.predict(x_train)
```

```python
y_pred_train
```

```python
classifier.score(x_train,y_train)
```

```python
classifier.score(x_test,y_test)
```

```python
y_pred_test = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred_test)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_test))
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(classifier,x,y,cv=5)
scores
```
